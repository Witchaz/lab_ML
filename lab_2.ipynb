```python
import pandas as pd
data = pd.read_csv("complaint.csv",index_col=0)
data
```

```python
data = data.dropna()
data
```

```python
data.columns = ["text","category","region","organization"]
data
```

```python
data.to_csv("complaint_fixed.csv",encoding="utf-8-sig",index=False)
```

```python
train_data,test_data = train_test_split(data,test_size=0.3,random_state=1234,stratify=data["category"])
train_data
```

```python
from imblearn.over_sampling import RandomOverSampler
```

```python
ros = RandomOverSampler(random_state=123456)
train_X , train_y = ros.fit_resample(train_data,train_data["category"])
train_X["category"].value_counts()
```

```python
from imblearn.under_sampling import RandomUnderSampler
```

```python
rus = RandomUnderSampler()
train_X,train_y = rus.fit_resample(train_data,train_data["organization"])
train_X["organization"].value_counts()
```
